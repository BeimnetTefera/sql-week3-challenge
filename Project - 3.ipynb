# B. Data Analysis Questions

## Q1. How many customers has Foodie-Fi ever had?

```sql
SELECT 
  COUNT(distinct customer_id) AS total_subs
FROM subscriptions;
```

## Q2. What is the monthly distribution of trial plan start_date values for our dataset - use the start of the month as the group by value?

```sql
SELECT 
  COUNT(*) AS monthly_sub_for_trial,
  DATE_FORMAT(sub.start_date, 'MMM') AS month,
  YEAR(sub.start_date) AS year
FROM subscriptions AS sub
JOIN plans AS pl
  ON pl.plan_id = sub.plan_id
WHERE pl.plan_name = 'trial'
GROUP BY 
    YEAR(sub.start_date),
    MONTH(sub.start_date),
    DATE_FORMAT(sub.start_date, 'MMM')
ORDER BY
    YEAR(sub.start_date),
    MONTH(sub.start_date)
```

## Q3. What plan start_date values occur after the year 2020 for our dataset? Show the breakdown by count of events for each plan_name?

```sql
SELECT 
  pl.plan_name,
  COUNT(*) AS subscription_after_2020
FROM workspace.foodie_fi.subscriptions AS sub
JOIN workspace.foodie_fi.plans AS pl
  ON pl.plan_id = sub.plan_id
WHERE YEAR(sub.start_date) > 2020
GROUP BY pl.plan_name;
```

## Q4. What is the customer count and percentage of customers who have churned rounded to 1 decimal place?

```sql
WITH base AS (
    SELECT 
      pl.plan_name,
      COUNT(*) AS total_subscriptions_per_plan
    FROM subscriptions AS sub
    JOIN plans AS pl
      ON pl.plan_id = sub.plan_id
    GROUP BY pl.plan_name
)
SELECT 
  SUM(
    CASE
      WHEN plan_name = 'churn' THEN total_subscriptions_per_plan
      ELSE 0
    END
    ) AS total_churn,
    
  ROUND(
    SUM(
      CASE
        WHEN plan_name = 'churn' THEN total_subscriptions_per_plan
        ELSE 0
      END
    ) / SUM(total_subscriptions_per_plan) * 100, 
  1) AS percentage_churn


FROM base;
```

## Q5. How many customers have churned straight after their initial free trial - what percentage is this rounded to the nearest whole number?

```sql
WITH first_plan AS (
  -- find each customers first plan
  SELECT 
    sub.customer_id,
    sub.start_date,
    pl.plan_name,
    ROW_NUMBER() OVER (PARTITION BY sub.customer_id ORDER BY sub.start_date) AS rnk
  FROM subscriptions AS sub
  JOIN plans AS pl
    ON pl.plan_id = sub.plan_id
),

trial_customers AS (
  -- only customers whose first plan was 'trial'
  SELECT *
  FROM first_plan
  WHERE plan_name = 'trial' and rnk = 1
),

churn_after_trial AS (
-- find customers who churned after trial
  SELECT DISTINCT t.customer_id
  FROM trial_customers AS t
  JOIN first_plan AS fp
      ON t.customer_id = fp.customer_id
  WHERE fp.plan_name = 'churn'
)
SELECT 
  COUNT(*) AS churned_after_trial,
  ROUND(
      COUNT(*) * 100 / (SELECT COUNT(*) FROM trial_customers),
  0) AS churned_after_trial_percent
FROM churn_after_trial;
```

## Q6. What is the number and percentage of customer plans after their initial free trial?

```sql
WITH ranked_plan(
    -- Rank each customers plans by start date
    SELECT 
        sub.customer_id,
        sub.start_date,
        pl.plan_name,
        ROW_NUMBER() OVER( PARTITION BY sub.customer_id ORDER BY sub.start_date) AS rnk
    FROM subscriptions AS sub
    JOIN plans AS pl
      ON pl.plan_id = sub.plan_id
),
first_trial_plan AS (
  -- Get customers whose first plan is a trial
  SELECT 
    customer_id,
    plan_name,
    rnk AS rnk_1
  FROM ranked_plan 
  WHERE plan_name = 'trial' AND rnk = 1
),
second_plan AS (
  -- Get each customers second plan
  SELECT 
    customer_id,
    plan_name,
    rnk AS rnk_2
  FROM ranked_plan
  WHERE rnk = 2
),

plan_after_trial AS (
  -- Combine customers trial plans with their next plans
  SELECT 
    ftp.customer_id,
    ftp.plan_name,
    sp.plan_name AS second_plan,
    ftp.rnk_1,
    sp.rnk_2
  FROM first_trial_plan AS ftp
  LEFT JOIN second_plan AS sp
    ON ftp.customer_id = sp.customer_id
)
SELECT 
  second_plan,
  COUNT(*) AS total_people_per_plan_after_trial,
  ROUND((COUNT(*) * 100.0 / SUM(COUNT(*)) OVER()), 1) AS percentage
FROM plan_after_trial
GROUP BY second_plan;
```

## Q7. What is the customer count and percentage breakdown of all 5 plan_name values at 2020-12-31?

```sql
WITH latest_plan AS (
  SELECT 
    sub.customer_id,
    pl.plan_name,
    sub.start_date,
    ROW_NUMBER() OVER (
        PARTITION BY sub.customer_id
        ORDER BY sub.start_date DESC
    ) AS rnk
  FROM subscriptions AS sub
  JOIN plans AS pl
    ON pl.plan_id = sub.plan_id
  WHERE sub.start_date <= '2020-12-31'
)
SELECT 
  plan_name,
  COUNT(*) AS customer_count,
  ROUND( 
        COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),
      2) AS percentage
FROM latest_plan
WHERE rnk = 1
GROUP BY plan_name;
```

## Q8. How many customers have upgraded to an annual plan in 2020?

```sql
WITH customer_plan_transitions AS (
  -- customers current plan alongside their previous plan
    SELECT 
        sub.customer_id,
        pl.plan_name AS current_plan,
        sub.start_date,
        LAG(pl.plan_name) OVER (
            PARTITION BY sub.customer_id 
            ORDER BY sub.start_date
        ) AS prev_plan
    FROM subscriptions AS sub
    JOIN plans AS pl
      ON pl.plan_id = sub.plan_id
)
SELECT 
  customer_id,
  prev_plan,
  current_plan
FROM customer_plan_transitions
WHERE current_plan = 'pro annual' AND
      prev_plan IS NOT NULL AND
      prev_plan <> 'pro annual' AND
      start_date BETWEEN '2020-01-01' AND '2020-12-31';
```

## Q9. How many days on average does it take for a customer to an annual plan from the day they join Foodie-Fi?

```sql
WITH customer_timeline AS(
  -- For every subscription event, attach the customers join date
    SELECT 
    sub.customer_id,
    sub.start_date,
    pl.plan_name,
    MIN(sub.start_date) OVER(PARTITION BY sub.customer_id)AS join_date
  FROM subscriptions AS sub
  JOIN plans AS pl
    ON pl.plan_id = sub.plan_id
),
first_annual_upgrade AS(
  -- For customers who reached the annual plan, find the first time they did so
  SELECT
    customer_id,
    plan_name,
    join_date,
    MIN(start_date) OVER(PARTITION BY customer_id) AS upgrade_annual
  FROM customer_timeline
  WHERE plan_name = 'pro annual'
)
-- Compute the average number of days from join to first annual upgrade
SELECT 
  ROUND(
      AVG(
        DATE_DIFF(day, join_date, upgrade_annual)
      ),
    2) AS avg_date_to_upgrade_annual
FROM first_annual_upgrade;
```

## Q10. Can you further breakdown this average value into 30 day periods (i.e. 0-30 days, 31-60 days etc)

```sql
WITH customer_timeline AS(
  -- For every subscription event, attach the customers join date
    SELECT 
    sub.customer_id,
    sub.start_date,
    pl.plan_name,
    MIN(sub.start_date) OVER(PARTITION BY sub.customer_id)AS join_date
  FROM subscriptions AS sub
  JOIN plans AS pl
    ON pl.plan_id = sub.plan_id
),
first_annual_upgrade AS(
  -- For customers who reached the annual plan, find the first time they did so
  SELECT
    customer_id,
    plan_name,
    join_date,
    MIN(start_date) OVER(PARTITION BY customer_id) AS upgrade_annual
  FROM customer_timeline
  WHERE plan_name = 'pro annual'
),
days_to_upgrade AS (
  -- Compute number of days from join to first annual upgrade
    SELECT 
      customer_id,
      DATE_DIFF(DAY, join_date, upgrade_annual) AS days_took
    FROM first_annual_upgrade
),
bucketed AS (
  -- Bucket customers into 30-day intervals
  SELECT
    customer_id,
    days_took,
    CASE
      WHEN days_took BETWEEN 0 AND 30 THEN 1
      WHEN days_took BETWEEN 31 AND 60 THEN 2
      WHEN days_took BETWEEN 61 AND 90 THEN 3
      WHEN days_took BETWEEN 91 AND 120 THEN 4
      ELSE 5
    END AS bucket_order,
    CASE
      WHEN days_took BETWEEN 0 AND 30 THEN '0-30 Days'
      WHEN days_took BETWEEN 31 AND 60 THEN '31-60 Days'
      WHEN days_took BETWEEN 61 AND 90 THEN '61-90 Days'
      WHEN days_took BETWEEN 91 AND 120 THEN '91-120 Days'
      ELSE '120+ Days'
    END AS bucket_label
  FROM days_to_upgrade
)
SELECT 
  bucket_label,
  COUNT(*) AS customer_count
FROM bucketed
GROUP BY bucket_label, bucket_order
ORDER BY bucket_order;
```

## Q11. How many customers downgraded from a pro monthly to a basic monthly plan in 2020?

```sql
WITH base AS (
  -- each customers subscription timeline and the next plan
    SELECT 
      sub.customer_id,
      sub.start_date,
      pl.plan_name,
      LEAD(pl.plan_name) OVER (
        PARTITION BY sub.customer_id
        ORDER BY start_date 
      ) AS next_plan
    FROM subscriptions AS sub
    JOIN plans AS pl
      ON pl.plan_id = sub.plan_id
)

SELECT 
    COUNT(*) AS customer_count
FROM base
WHERE plan_name = 'pro monthly'
  AND next_plan = 'basic monthly'
  AND YEAR(start_date) = 2020
```
